```python
# import dataset, sample 30 random interactions

def sample_validation_interactions(import_path, export_path, subreddit, random_seed = 42, n = 30):

    if subreddit == "politics":
        # read from zip
        df_chunks = []
        with zipfile.ZipFile(import_path, 'r') as zf:
            file_list = zf.namelist()  # Get all file names in ZIP
            file_list.sort()  
        
            print("Files found in ZIP:", file_list)  
        
            for file in file_list:
                with zf.open(file) as f:
                    df_chunk = pd.read_csv(f,  quotechar='"', encoding="utf-8")  
                    df_chunks.append(df_chunk) 
        
        
        df = pd.concat(df_chunks, ignore_index=True)

    else:
        df = pd.read_csv(import_path)


    sample = df.sample(n = n, random_state = random_seed).reset_index(drop = True)
    #sample.to_csv(export_path, index = False)

    return sample

```
