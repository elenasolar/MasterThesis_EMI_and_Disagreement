# **Get Unique Users**

```python
import pandas as pd
import zipfile
```

```python
# import files and extract unique users

def extract_unique_users(subreddit):

    filepath = f"../preprocessing/{subreddit}/output/{subreddit}_new_preprocessing_com_rep.csv"
    data = pd.read_csv(filepath)

    authors = set(data['author_child']).union(data['author_parent'], data['author_submission'])

    return list(authors)
```

```python
# run on all subreddits

subreddits = ["AmericanPolitics", "Ask_Politics", "PoliticalDebate", "PoliticalDiscussion"]
users_dict = {}

for subr in subreddits:
    users_dict[subr] = extract_unique_users(subr)
```

```python
# count and concat

print(len(users_dict["Ask_Politics"]))
print(len(users_dict["PoliticalDebate"]))
print(len(users_dict["PoliticalDiscussion"]))
print(len(users_dict["AmericanPolitics"]))

print("Sum of all: ", len(users_dict["Ask_Politics"] + users_dict["PoliticalDebate"] + users_dict["PoliticalDiscussion"] + users_dict["AmericanPolitics"]))
print("Sum of unique: ", len(set(users_dict["Ask_Politics"] + users_dict["PoliticalDebate"] + users_dict["PoliticalDiscussion"] + users_dict["AmericanPolitics"])))
```
